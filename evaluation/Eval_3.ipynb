```python
def generate_prompt(evaluation_criteria, evaluation_steps, json_schema, examples, question, ground_truth,answer):
    prompt = f'''
You are an expert judge evaluating the Retrieval Augmented Generation applications. 
Your task is to evaluate a given answer based on a ground truth and question using the criteria provided below.
 
Evaluation Criteria (0-1):
{evaluation_criteria}
 
Evaluation Steps:
{evaluation_steps}
 
Output format:
{json_schema}

Examples:
{examples}
 
Now, please evaluate the following:
 
Question:
{question}
Ground Truth:
{ground_truth}
Answer:
{answer}
'''
    return prompt
```

```python
import json

class Data:
    def read_jsonl_file(self, file_path):
        data = []
        with open(file_path, 'r') as file:
            for line in file:
                data.append(json.loads(line))
        return data

    def save_result(self, resultList):
        with open(f'result.json', 'w') as output_json_file:
            json.dump({"results": resultList}, output_json_file)

        print(f"\Evaluations saved to result.json")
```

```python
from tqdm.auto import tqdm
import json
def main(inputFile, test = False):
        # funciton variables
        resultList = []
        
        # load evaluation dataset
        dataloader = Data()
        dataset = dataloader.read_jsonl_file(inputFile)
        
        for data in tqdm(dataset):
            print('Example: ' + data['id'])
            # generate the prompt
            answer = data['response']
            ground_truth = data['ground_truth']
            question = "What kind of animal is that? Be as specific as possible! Fish and bird for example are too coarse."
            prompt = generate_prompt(evaluation_criteria_simple, evaluation_steps_simple, json_schema_simple, examples_simple, question, ground_truth, answer)

            #print(prompt)
            # run model
            evaluation = pipe(prompt_template(prompt))[0]['generated_text']
            print(evaluation)
            
            # store result
            resultDict = {}
            resultDict['model'] = 'Llama-2'
            resultDict['id'] = data['id']
            resultDict['input'] = data['response']
            resultDict['ground_truth'] = data['ground_truth']
            resultDict['evaluation'] = evaluation
            resultList.append(resultDict)

            if test == True:
                break

        # save the results in a json file
        dataloader.save_result(resultList)
```

```python
import numpy as np
def calculate_rmse(component_scores, human_scores, component):
        errors = []
        for model_score, human_score in zip(component_scores, human_scores):
            error = (model_score[component] - human_score[component]) ** 2
            errors.append(error)
        mse = np.mean(errors)
        rmse = np.sqrt(mse)
        return rmse
```

## Bias analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, ttest_ind

# Example data
data = {
    'id': range(1, 26),
    'human_scores': [3, 3, 6, 4, 4, 4, 3, 3, 5, 0, 3, 3, 3, 3, 3, 3, 3, 3, 3, 12, 3, 8, 1, 3, 4],
    'phi_scores': [1, 3, 6, 4, 4, 6, 4, 3, 6, 1, 3, 3, 4, 3, 3, 4, 3, 4, 3, 12, 3, 8, 3, 3, 4],
    'hermes_scores': [3, 3, 5, 3, 6, 8, 11, 3, 3, 2, 3, 3, 3, 3, 3, 3, 3, 5, 3, 13, 2, 10, 3, 2, 9],
    'group': ['A', 'A', 'B', 'B', 'A', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# Calculate correlations
phi_human_corr = pearsonr(df['phi_scores'], df['human_scores'])[0]
hermes_human_corr = pearsonr(df['hermes_scores'], df['human_scores'])[0]

print(f'Pearson correlation between human and phi scores: {phi_human_corr:.2f}')
print(f'Pearson correlation between human and hermes scores: {hermes_human_corr:.2f}')

# Compare score distributions
plt.figure(figsize=(10, 6))
df[['human_scores', 'phi_scores', 'hermes_scores']].boxplot()
plt.title('Score Distributions')
plt.ylabel('Scores')
plt.show()

# Hypothesis testing for group differences
group_a_phi = df[df['group'] == 'A']['phi_scores']
group_b_phi = df[df['group'] == 'B']['phi_scores']
t_stat, p_value = ttest_ind(group_a_phi, group_b_phi)
print(f'T-test between groups for phi scores: t-statistic={t_stat:.2f}, p-value={p_value:.2f}')

group_a_hermes = df[df['group'] == 'A']['hermes_scores']
group_b_hermes = df[df['group'] == 'B']['hermes_scores']
t_stat, p_value = ttest_ind(group_a_hermes, group_b_hermes)
print(f'T-test between groups for hermes scores: t-statistic={t_stat:.2f}, p-value={p_value:.2f}')

# Error analysis
df['phi_error'] = df['phi_scores'] - df['human_scores']
df['hermes_error'] = df['hermes_scores'] - df['human_scores']

# Visualize errors
plt.figure(figsize=(10, 6))
df[['phi_error', 'hermes_error']].boxplot()
plt.title('Error Distributions')
plt.ylabel('Error')
plt.show()

# Group-based error analysis
group_error_means = df.groupby('group')[['phi_error', 'hermes_error']].mean()
print(group_error_means)
```
